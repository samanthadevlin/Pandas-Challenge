```python
#Purchase Count
gender_count=purchase_data[["Price","Gender"]].groupby("Gender").count()
gender_count
```

```python
#Average Purchase Price 
average_purch_price=round(purchase_data[["Price","Gender"]].groupby("Gender").mean(),2)
average_purch_price
```

```python
#Total Purchase Value 
total_purch_price=purchase_data[["Price","Gender"]].groupby("Gender").sum()
total_purch_price
```

```python
#used group by function and aggregrate function 
#sources used https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html and https://pandas.pydata.org/pandas-docs/version/0.24.2/reference/api/pandas.DataFrame.agg.html
avePurch=purchase_data.groupby(["Gender"])["Price"].agg(["count","mean","sum"])
avePurch
#used .columns funcation to rename columns 
avePurch.columns=["Purchase Count", "Average Purchase Price","Avg Total Purchase Per Person",]
avePurch
```

```python
purchasecount=round(avePurch.loc[:,"Purchase Count"],2)
purchasecount 
```

```python
ave_purch_price=round(avePurch.loc[:,"Average Purchase Price"],2)
ave_purch_price
```

```python
amount_purch_gender=round(purchasecount.div(count),2)
amount_purch_gender
```

```python
total_purch_value=round(amount_purch_gender.mul(ave_purch_price),2)
total_purch_value
```

```python
avePurch.insert(2,"Total Purchase Value", total_purch_value, allow_duplicates = False)
avePurch
avePurch["Average Purchase Price"] = avePurch["Average Purchase Price"].map("${:,.2f}".format)
avePurch["Total Purchase Value"] = avePurch["Total Purchase Value"].map("${:,.2f}".format)
avePurch["Avg Total Purchase Per Person"] = avePurch["Avg Total Purchase Per Person"].map("${:,.2f}".format)
avePurch
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
#established bins for ages 
age_bins=[0, 9, 14, 19, 24, 29,34,39,999]
age_group=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["AgeRanges"]=pd.cut(purchase_data["Age"],age_bins, labels=age_group)
age_group=purchase_data.groupby("AgeRanges")
age_group.first()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby to get total Screen name 
top_spender_group=purchase_data.groupby(["SN"])
top_spender_group.first()
```

```python
#used count and mean functions to pull specific data needed to find top sepnders and top sepender in group
top_spender_count=top_spender_group["Purchase ID"].count()
top_spender_count
top_spender_mean=top_spender_group["Price"].mean()
top_spender_mean
#total spender average
top_spender_ave=top_spender_count*top_spender_mean
top_spender_ave
```

```python
#created a data frame
top_spender_df=pd.DataFrame({"Purchase Count":top_spender_count, "Average Purchase Price":top_spender_mean,"Total Purchase Value":top_spender_ave})
top_spender_df=top_spender_df.sort_values('Total Purchase Value',ascending=False)
top_spender_df
```
